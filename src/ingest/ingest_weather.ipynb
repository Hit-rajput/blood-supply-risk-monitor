```python
weekly_weather = df_weather.resample('W-MON').agg({
    'prcp': 'sum',   
    'snow': 'max',   
    'tmin': 'min'    
})
```

```python
weekly_weather['snow_depth_cm'] = weekly_weather['snow'] / 10.0
weekly_weather.rename(columns={
    'prcp': 'precip_mm',
    'tmin': 'min_temp'
}, inplace=True)
```

```python
weekly_weather = weekly_weather[['precip_mm', 'snow_depth_cm', 'min_temp']]
```
