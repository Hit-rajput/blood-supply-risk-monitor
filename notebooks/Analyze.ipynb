```python
print("\n" + "="*80)
print("TORONTO DATASET ANALYSIS")
print("="*80)
print(f"Shape: {df_toronto.shape}")
print(f"\nColumns: {list(df_toronto.columns)}")
print(f"\nData types:\n{df_toronto.dtypes}")
print(f"\nYear range: {df_toronto['OCC_YEAR'].min()} - {df_toronto['OCC_YEAR'].max()}")
```

```python
print(f"\nFATALITIES summary:")
print(df_toronto['FATALITIES'].describe())
```

```python
print("\n" + "="*80)
print("OVERLAP PERIOD ANALYSIS (2014-2021)")
print("="*80)

ncdb_overlap = df_ncdb[df_ncdb['C_YEAR'].between(2014, 2021)]
toronto_overlap = df_toronto[df_toronto['OCC_YEAR'].between(2014, 2021)]

print(f"NCDB records (2014-2021): {len(ncdb_overlap):,}")
print(f"Toronto records (2014-2021): {len(toronto_overlap):,}")
```

```python
print("\nYearly collision counts:")
ncdb_yearly = ncdb_overlap.groupby('C_YEAR').size()
toronto_yearly = toronto_overlap.groupby('OCC_YEAR').size()

comparison = pd.DataFrame({
    'Year': ncdb_yearly.index,
    'NCDB_National': ncdb_yearly.values,
    'Toronto': toronto_yearly.values
})
comparison['Toronto_Share'] = (comparison['Toronto'] / comparison['NCDB_National'] * 100).round(2)
print(comparison)
```
